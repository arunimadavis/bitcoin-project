```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
coin=pd.read_csv('BitcoinPrice.csv')
```

```python
coin
```

```python
coin.info()
```

```python
predictdays=30
```

```python
coin['prediction']=coin[['Price']].shift(-predictdays)
```

```python
coin.head()
```

```python
x=np.array(coin.drop(columns=['prediction'],axis=1))
```

```python
x=x[:len(coin)-predictdays]
```

```python
x.shape
```

```python
x
```

```python
y=np.array(coin['prediction'])
```

```python
y=y[:-predictdays]
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
predictarray=np.array(coin.drop(columns=['prediction'],axis=1))[-predictdays:]
```

```python
print(predictarray)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
rfpredict=rf.predict(x_test)
```

```python
rfpredict
```

```python
print('random accuracy: {: .2f}%'.format(rf.score(x_test,y_test)*100))
```
